```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
import seaborn as sns

race_data = pd.read_excel(r".\\CSV Files\\Combined data.xlsx")
#print(race_data['Generation'][race_data['Generation'] != "Japanese"].shape)
#print(race_data['Generation'].dropna().shape)
race_data['Generation'] = race_data['Generation'][(race_data['Generation'] != "Japanese") & (race_data['Generation'] != 'Jewish')] 

#race_data['Generation'] = race_data['Generation'].dropna()
race_data = race_data.dropna(axis=0, subset=['Generation'])
#print(race_data['Generation'].isnull().sum())
#print(race_data[race_data['Generation'] != "japanese"].shape)
race_data['Generation'][race_data['Generation'] == 'Ni-ten-gosei (half Nisei, half Sansei)'] = 'Ni-ten-gosei'
race_data['Generation'][race_data['Generation'] == 'White'] = 'White American'
#race_data['Generation'][race_data['Generation'] == 'Jewish']
print(race_data.shape)
#race_data['Generation'] = race_data['Generation'].map({'Ni-ten-gosei (half Nisei, half Sansei)': 'Ni-ten-gosei'})
plt.rcParams['figure.figsize'] = [12, 10]
ax = sns.countplot(x='Generation', data=race_data)
ax.bar_label(ax.containers[0])
ax.text(5.6, 570, "Total sample count is {}".format(race_data['Generation'].shape[0]),fontsize=12)
num = [0, 2, 3, 4, 5]
for i, bar in enumerate(list([ax.patches[0] , ax.patches[2], ax.patches[3], ax.patches[4], ax.patches[5], ax.patches[6], ax.patches[7]])):
    bar.set_color("blue")
bar.set_label('Chinese American')

for i, bar in enumerate(list([ax.patches[1]])):
    bar.set_color("magenta")
bar.set_label('White American')


plt.title("Narrators Race Analysis")
plt.xlabel("Races of Narrators")
plt.ylabel("Count of Races")
plt.legend(loc='best')

plt.xticks(rotation=90)
plt.show()
```
