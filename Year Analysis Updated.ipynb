```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import pandas as pd
#import csv
year_data = pd.read_csv(r".\\CSV Files\\Year_of_Narrator.csv")
#data.head()
#data.info()
year_data = year_data.dropna() 
year_data = year_data[(year_data['Year'] > 1800) & (year_data['Year'] < 2022)]
#data[["Year"]].astype(int64)
year_data['Year'] = year_data['Year'].astype('int64').astype('str')

year_data = year_data.sort_values(by=['Year'], ascending=True)
print(year_data.head())
```

```python
year_data['Year'] = year_data['Year'].astype('int64')
print(year_data.info())
```

```python
import seaborn as sns
plt.rcParams['figure.figsize'] = [30, 15]
ax = sns.countplot(x='Year', data=year_data)
ax.bar_label(ax.containers[0])
plt.xlabel("Year of Birth of Narrator")
plt.ylabel("Count of Years")
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns

plt.rcParams['figure.figsize'] = [8, 6]
ax = sns.barplot(x='Birth year', y = 'Total number', data=df)
#ax.set_label(df['label'].values)

ax.bar_label(ax.containers[0])
ax.text(6.5, 240, "Total sample count is {}".format(df['Total number'].sum(), fontsize=10))

for i, bar in enumerate(ax.patches[:2]):
    #bar.set_color("C{}".format(i))
    bar.set_color("magenta")
bar.set_label('Not born yet in 1942')

for i, bar in enumerate(ax.patches[2:5]):
    #bar.set_color("C{}".format(i))
    bar.set_color("yellow")
bar.set_label('Minor in 1942')
    
for i, bar in enumerate(ax.patches[5:]):
    #bar.set_color("C{}".format(i))
    bar.set_color("cyan")
bar.set_label('Adult in 1942')

plt.title("Analysis of Birth Year")
plt.xlabel("Narrators Birth Year Distribution")
plt.ylabel("Count of Narrators")     
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.legend(loc='best')
plt.show()
```
